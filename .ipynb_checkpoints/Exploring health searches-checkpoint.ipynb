```python
import pandas as pd
import matplotlib.pyplot as mpl
import seaborn as sns

data = pd.read_csv('health_searches.csv')
data.head()
```

```python
def render_heatmap(data, title):
    data_corr = data.corr()
    sns.heatmap(data_corr, annot = True, xticklabels = data_corr.columns, yticklabels = data_corr.columns)
    mpl.suptitle(title)
    mpl.show()
    
def render_scattermatrix(data, title, xsize, ysize):
    pd.plotting.scatter_matrix(data, figsize = (xsize, ysize))
    mpl.tight_layout()
    mpl.suptitle(title)
    mpl.show()
    
search_topics = ['cancer', 'cardiovascular', 'stroke', 'depression', 'rehab', 'vaccine', 'diarrhea', 'obesity', 'diabetes']
```

```python
# %% desccriptive analytics
data.describe()
```

```python
# %% viewing overall data distributions
data.hist(bins = 30, figsize = (40, 35))
mpl.suptitle('Fig.1: Overall data distribution histograms')
mpl.show()
```

```python
i = 0
for search_str in search_topics: 
    i += 1
    curr = data.filter(regex = search_str)
    curr.hist(bins = 30)
    mpl.suptitle('Fig 1.' + str(i) + ': ' + search_str + ' search histograms')
    mpl.show()
    print(curr.describe())
```

```python
i = 0
for search_str in search_topics: 
    i += 1
    curr = data.filter(regex = search_str)
    curr.boxplot()
    mpl.suptitle('Fig 2.' + str(i) + ': ' +search_str + ' search boxplots')
    mpl.show()
```

```python
# data.plot.scatter(x = '2004+cardiovascular', y = '2004+diarrhea')
i = 0
for search_topic in search_topics: 
    i += 1
    curr = data.filter(regex = search_topic)
    render_heatmap(curr, 'Fig. 3.'  + str(i) + ': ' + ': Diabetes heatmap')
```

```python
correlations = data.corr()
correlations.head()
```

```python
# %% view correlations per year 
i = 0
for year in range(2004, 2018): 
    curr = data.filter(regex = str(year))
    i += 1
    render_heatmap(curr, 'Fig 4.' + str(i) + ': ' + str(year) + ' searches correlation heatmap')
```

```python
# %% viewing scatterplot for highest correlation in year 2005
# data_05 = data.filter(regex = '2005')
data_05 = pd.DataFrame(data, columns = ['dma', '2005+cancer', '2005+cardiovascular', '2005+stroke', '2005+depression', '2005+rehab', '2005+vaccine', '2005+diarrhea', '2005+obesity', '2005+diabetes'])
data_05.head()
# render_scattermatrix(data_05, '2005 scatterplot matrix', 20, 20)
# data_05.plot.scatter(x = '2005+depression', y = '2005+diabetes')

# print(data_05.groupby(['dma']))
sns.lmplot(x  = '2005+depression', y = '2005+diabetes', data = data_05) #~, hue = 'dma', ci = None , hue = 'dma'
mpl.suptitle('Fig. 5: 2005 scatterplot between depression and diabetes searches')
mpl.show()
```
