```python
import pandas as pd
import matplotlib.pyplot as mpl
import seaborn as sns

data = pd.read_csv('health_searches.csv')
data.head()
```

```python
data.describe()
```

```python
correlations = data.corr()
correlations.head()
```

```python
def render_heatmap(data, title):
    data_corr = data.corr()
    sns.heatmap(data_corr, annot = True, xticklabels = data_corr.columns, yticklabels = data_corr.columns)
    mpl.suptitle(title)
    mpl.show()
```

```python
# %% ~explore 2004 searches
searches_04 = pd.DataFrame(data, columns = ['2004+cancer', '2004+cardiovascular', '2004+stroke', '2004+depression', '2004+rehab', '2004+vaccine', '2004+diarrhea', '2004+obesity', '2004+diabetes'])
render_heatmap(searches_04, '2004 searches correlation heatmap')

# %% highest corr, diabetes and depression
searches_05 = pd.DataFrame(data, columns = ['2005+cancer', '2005+cardiovascular', '2005+stroke', '2005+depression', '2005+rehab', '2005+vaccine', '2005+diarrhea', '2005+obesity', '2005+diabetes'])
render_heatmap(searches_05, '2005 searches correlation heatmap')
```

```python
# %% filtering by year
filter_17 = data.filter(regex='2017')
filter_17.head()
```

```python
# %% filtering by search str, diabetes
filter_diabetes = data.filter(regex='diabetes')
filter_diabetes.head()
diabetes_corr = filter_diabetes.corr()
sns.heatmap(diabetes_corr, annot = True, xticklabels=diabetes_corr.columns, yticklabels=diabetes_corr.columns)
mpl.show()
```
