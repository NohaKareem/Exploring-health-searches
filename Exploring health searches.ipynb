```python
import pandas as pd
import matplotlib.pyplot as mpl
import seaborn as sns

data = pd.read_csv('health_searches.csv')
data.head()
```

```python
def render_heatmap(data, title):
    data_corr = data.corr()
    sns.heatmap(data_corr, annot = True, xticklabels = data_corr.columns, yticklabels = data_corr.columns)
    mpl.suptitle(title)
    mpl.show()
    
def render_scattermatrix(data, title, xsize, ysize):
    pd.plotting.scatter_matrix(data, figsize = (xsize, ysize))
    mpl.tight_layout()
    mpl.suptitle(title)
    mpl.show()
    
search_topics = ['cancer', 'cardiovascular', 'stroke', 'depression', 'rehab', 'vaccine', 'diarrhea', 'obesity', 'diabetes']
```
